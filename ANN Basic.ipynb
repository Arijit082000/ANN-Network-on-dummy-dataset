```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Creating a dummy sataset with 1000 sample and 10 features
x,y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

```python
#Split dataset into 80% training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
#Scaling dataset using StandardScaler
# Scaling dataset using StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Here we will use the Sequential model, where the data will pass through one layer after another.
model = Sequential()
```

```python
# Input layer and first hidden layer (16 neurons, ReLU activation)
# input_shape=(10,) because we have 10 features
model.add(Dense(16, activation='relu', input_shape=(10,)))
```

```python
# Second hidden layer (8 neurons, ReLU activation)
model.add(Dense(8, activation='relu'))
```

```python
# Output layer (1 neuron, Sigmoid activation)
# Since this is binary classification (output 0 or 1), sigmoid is used
model.add(Dense(1, activation='sigmoid'))
```

```python
# View the model summary
model.summary()
```

```python
# Compiling the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Training the model (Epochs = 20)
historry = model.fit(x_train, y_train, epochs=20, batch_size=32, validation_split=0.2)
```

```python
# Check model accuracy on test data
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
